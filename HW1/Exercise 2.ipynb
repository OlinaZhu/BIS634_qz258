```python
#importing libraries and data
import matplotlib
import pandas as pd
import sqlite3
with sqlite3.connect("hw1-population (1).db") as db:
    data = pd.read_sql_query("SELECT * FROM population", db)
data.head()
```

```python
# columns names and number of rows
print("The columns are:", ", ".join(list(data.columns)))
print("There are", data['name'].count(), "rows in the data")
```

```python
# distribution of ages
data['age'].describe()
```

```python
# histogram of age distribution
age_hist = data['age'].plot.hist(bins=20)
age_hist.grid(color='white',lw=0.5, axis='x')
```

```python
# distribution of weights
data['weight'].describe()
```

```python
# histogram of weight distribution
weight_hist = data['weight'].plot.hist(color='green', bins=15)
weight_hist.grid(color='white',lw=0.5, axis='x')
```

```python
# scatterplot of weight vs age
weight_age = data[['weight','age']].plot.scatter(x='age', y='weight', s=0.5)
```

```python
# weight vs age outlier name
over_forty = data.loc[data['age']>40]
outlier_info = over_forty.loc[data['weight']<30]
outlier_name = "".join(outlier_info['name'])
print("The outlier's name is", outlier_name)
```
